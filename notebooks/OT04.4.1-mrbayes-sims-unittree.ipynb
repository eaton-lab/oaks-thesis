Tree with variation in Ne
=====

```python
# Reduce to 30-40 tips?
treuv = toytree.rtree.unittree(200, treeheight = 5.6e7, seed = 234)
treuv.draw(layout = "d", scalebar = True);
```

```python
# Create a dictionary of Ne values, chosen randomly from a given interval, then set onto the tree.
dict_ne = {i.name : random.randint(1e4, 2e4) for i in treuv.get_feature_dict()}
treuv = treuv.set_node_values("Ne", dict_ne)
```

```python
# We will assume gentime = 10 across the tree to start.
dict_g = {i.name : random.randint(8, 12) for i in treuv.get_feature_dict()}
treuv = treuv.set_node_values("g", dict_g)
```

```python
# Divide edge lengths (absolute time) by generation time to get units of generations.
ttreuv = treuv.set_node_values(
    "dist",
    {i.name: i.dist / i.g for i in treuv.get_feature_dict()}
)
```

```python
ttreuv.draw(layout = "d", scalebar = True);
```

```python
moduv = ipcoal.Model(ttreuv, nsamples = 2)
moduv.sim_loci(10, 1000)
```

```python
moduv.write_concat_to_nexus(name="mbtest-uv", outdir='/moto/eaton/users/hnl2109/analysis-mrbayes/', diploid=True)
```

```python
moduv.df
```

```python
mbuv2 = ipa.mrbayes(
    data="/moto/eaton/users/hnl2109/analysis-mrbayes/mbtest-uv.nex",
    name="itest-uv2",
    workdir="/moto/eaton/users/hnl2109/analysis-mrbayes/",
    clock_model=2,
    constraints=ttreuv,
    ngen=int(5e6),
    nruns=2,
)

mbuv2.params.samplefreq = 200

# Print parameters.
print(mbuv2.params)
```

```python
mbuv2_2 = ipa.mrbayes(
    data="/moto/eaton/users/hnl2109/analysis-mrbayes/mbtest-uv.nex",
    name="itest-uv2-2",
    workdir="/moto/eaton/users/hnl2109/analysis-mrbayes/",
    clock_model=2,
    constraints=ttreuv,
    ngen=int(10e6),
    nruns=2,
)

mbuv2_2.params.samplefreq = 200

# Print parameters.
print(mbuv2_2.params)
```

```python
# Start ipyclient.
ipyclient = ipp.Client()
```

```python
ipyclient.ids
```

```python
# To do parallel runs, add block = False to the arguments, along with an ipyclient argument.
mbuv2.run(force = True, block = False, ipyclient = ipyclient)
mbuv2_2.run(force = True, block = False, ipyclient = ipyclient)
```

```python
mbuv2.convergence_stats
```
